```python
import yfinance as yf
import pandas as pd ;
import matplotlib.pyplot as plt
import numpy as np
```

```python
ticker = ["SPY", "AAPL", "KO", "IBM", "DIS", "MSFT", "AMZN", "GOOGL", "TSLA", "NFLX", "NVDA", "JPM", "V", "PG", "JNJ", "WMT", "BAC", "XOM", "T", "HD", "UNH", "PFE", "MRK", "NKE"]
```

```python
stocks = yf.download(ticker, start = "2012-01-01" , end = "2021-01-01" )
```

```python
stocks
```

```python
stocks.describe()
```

```python
close = stocks.loc[:,"Close"].copy()
close
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-darkgrid")
close.plot(figsize = (15 ,8) , fontsize = 12 )
plt.legend(fontsize = 16)
plt.show
```

```python
normclose = close.div(close.iloc[0]).mul(100)
normclose
```

```python
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")

normclose.plot(figsize = (12 ,8) , fontsize = 12 )
plt.legend(fontsize = 16)
plt.show
```

```python
ret = close.pct_change().dropna()
ret
```

```python
summary = close.describe().T.loc[:,["mean" , "std"]]
summary
```

```python
ret.mean()
```

```python
ret.var()
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] =  summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x = "std",y = "mean",figsize=(12 ,8 ),s=50,fontsize =15)
for i in summary.index :
            plt.annotate(i, xy=(summary.loc[i, "std"] + 0.5, summary.loc[i, "mean"] + 0.5), size=15)
plt.xlabel("Annual risk(std)" , fontsize = 15 )
plt.ylabel("Annual return" , fontsize = 15 )
plt.title("Risk/Return" , fontsize = 20 )
plt.show()
```
